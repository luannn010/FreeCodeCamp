```python
bachelors = (df['education'] == 'Bachelors').sum()

# Calculate the percentage of rows with 'Bachelors' education
percentage_bachelor = bachelors / len(df) * 100

print(percentage_bachelor)
```

```python
higher_education = df[df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]
lower_education = df[~df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]

# Calculate the percentage of people with advanced education and salary >50K
higher_education_rich = round((higher_education[higher_education['salary'] == '>50K'].shape[0] / higher_education.shape[0]) * 100, 1)
# Calculate the percentage of people with lower education and salary >50K
lower_education_rich = round((lower_education[lower_education['salary'] == '>50K'].shape[0] / lower_education.shape[0]) * 100, 1)
print(higher_education_rich)
print(lower_education_rich)
```
